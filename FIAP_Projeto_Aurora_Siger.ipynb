```python
import random
```

```python
# ── Faixas e regras ────────────────────────

def gerar_telemetria():
    iv = round(random.uniform(18, 35), 1)   # temperatura interna (°C)
    ev = round(random.uniform(-15, 30), 1)   # temperatura externa (°C)
    si = random.choice([0, 1])              # integridade estrutural (0/1)
    tp = round(random.uniform(95, 145), 2)  # pressão dos tanques (bar)
    cm = random.choice([0, 1])              # módulos críticos (0/1)

    # ── Dicionario com os dados da telemetria ──────────────

    dados_telemetria = {
        'internal_temp': iv,
        'external_temp': ev,
        'structural_integrity': si,
        'tank_pressure': tp,
        'critical_modules': cm,
    }
    # ── dicionario de status individual ─────────────────────────

    status_telemetria = {
        'internal_temp':        '[ OK ]'     if 21 <= iv <= 30  else '**FALHOU**',
        'external_temp':        '[ OK ]'     if -5 <= ev <= 30  else '**FALHOU**',
        'structural_integrity': '[ OK ]'     if si == 1         else '**FALHOU**',
        'tank_pressure':        '[ OK ]'     if 95 <= tp <= 145 else '**FALHOU**',
        'critical_modules':     '[ OK ]'     if cm == 1         else '**FALHOU**',
    }


    return dados_telemetria, status_telemetria
```

```python
def calcular_autonomia():
    capacidade_total  = round(random.uniform(800, 1200), 1)   # capacidade total da bateria
    carga_atual_percent   = round(random.uniform(60, 100), 1)   # carga atual (%)
    consumo_decolagem_kw  = round(random.uniform(15, 30), 1)    # consumo estimado na decolagem
    perda_energetica_perc = round(random.uniform(1, 7), 1)      # perdas energéticas (%)

    # ── Cálculos intermediários ───────────────────────────────────────────────
    energia_disponivel_kwh = round(capacidade_total * (carga_atual_percent / 100), 1)
    perda_kwh              = round(energia_disponivel_kwh * (perda_energetica_perc / 100), 2)
    energia_util_kwh       = round(energia_disponivel_kwh - perda_kwh, 4)
    autonomia_meses        = round(random.uniform(3, 15)) # autonomia estimada em meses
    autonomia_horas      = round(autonomia_meses * 30 * 24) # convertendo meses para horas

    dados_autonomia = {
        'capacidade_total':      capacidade_total,
        'energia_disponivel_kwh': energia_disponivel_kwh,
        'carga_atual_percent':   carga_atual_percent,
        'consumo_decolagem_kw':  consumo_decolagem_kw,
        'perda_energetica_perc': perda_energetica_perc,
        'energia_util_kwh':      energia_util_kwh,
        'autonomia_meses':     autonomia_meses,
        'autonomia_horas':  autonomia_horas,
    }

    status_autonomia = {
        'Bateria':    '[ OK ]'   if carga_atual_percent   >= 80  else 'INSUFICIENTE',
        'Energia necessária':   '[ OK ]'   if energia_util_kwh      >= 10  else 'INSUFICIENTE',
        'Perda energética':          '[ OK ]'    if perda_energetica_perc <= 5   else 'CRITICA',
        'Autonomia ':      '[ OK ]'   if autonomia_meses     >= 10  else 'INSUFICIENTE',
    }

    return dados_autonomia, status_autonomia
```

```python

# ── Usando as funcoes e retornando dados ─────────────────────────────────────────
dados_autonomia, status_autonomia = calcular_autonomia()
dados_telemetria, status_telemetria = gerar_telemetria()


 # ── Decisão de lançamento ──────────────────

decisao = 'PRONTO PARA DECOLAR' if all(
    s == '[ OK ]' for s in {**status_telemetria, **status_autonomia}.values()
) else 'DECOLAGEM ABORTADA'
```

```python

# --- imprimindo os resultados de forma organizada ------------------------

print("Iniciando Projeto Aurora-Siger")
print("...\n...")
print(input("Pressione Enter para iniciar as análises "))
print("Calculando autonomia da bateria...")
print(f"Processando {random.randint(25, 150)} calculos de autonomia...")
input("Pressione Enter para continuar...")
print("Gerando os calculos de telemetria...")
print(f"{random.randint(25, 150)} registros de telemetria processados...")
input("Pressione Enter para continuar...")

print("\n\n           === DADOS DE AUTONOMIA ===\n")
print(f"    Capacidade total da bateria:        {dados_autonomia['capacidade_total']}kW")
print(f"    Carga atual da bateria:             {dados_autonomia['energia_disponivel_kwh']}kW ({dados_autonomia['carga_atual_percent']}%)")
print(f"    Consumo estimado na decolagem:      {dados_autonomia['consumo_decolagem_kw']}kW")
print(f"    Perda energetica:                   {dados_autonomia['perda_energetica_perc']}%")
print(f"    Bateria apos decolagem:             {dados_autonomia['autonomia_horas']} horas ({dados_autonomia['autonomia_meses']} meses)")

print("\n\n           === DADOS DE TELEMETRIA ===\n")
print(f"    Temperatura interna:                {dados_telemetria['internal_temp']}ºC")
print(f"    Temperatura externa:                {dados_telemetria['external_temp']}ºC")
print(f"    Integridade estrutural:             {dados_telemetria['structural_integrity']}")
print(f"    Pressão dos tanques:                {dados_telemetria['tank_pressure']}bar")
print(f"    Módulos críticos:                   {dados_telemetria['critical_modules']}")

print("\n\n           === STATUS DE AUTONOMIA ===\n")
print(f"    Bateria:            {status_autonomia['Bateria']}")
print(f"    Energia necessária: {status_autonomia['Energia necessária']}")
print(f"    Perda energética:   {status_autonomia['Perda energética']}")
print(f"    Autonomia:          {status_autonomia['Autonomia ']}")

print("\n\n           === STATUS DE TELEMETRIA ===\n")
print(f"    Temperatura interna:        {status_telemetria['internal_temp']}")
print(f"    Temperatura externa:        {status_telemetria['external_temp']}")
print(f"    Integridade estrutural:     {status_telemetria['structural_integrity']}")
print(f"    Pressão dos tanques:        {status_telemetria['tank_pressure']}")
print(f"    Módulos críticos:           {status_telemetria['critical_modules']}")




print(f"\nDecisão de lançamento: {decisao}")
```
