```python
import random
```

```python
# ── Ranges and rules ────────────────────────

def generate_telemetry():
    iv = round(random.uniform(18, 35), 1)   # internal temperature (°C)
    ev = round(random.uniform(-15, 30), 1)   # external temperature (°C)
    si = random.choice([0, 1])              # structural integrity (0/1)
    tp = round(random.uniform(95, 145), 2)  # tank pressure (bar)
    cm = random.choice([0, 1])              # critical modules (0/1)

    # ── Dictionary with telemetry data ──────────────

    telemetry_data = {
        'internal_temp': iv,
        'external_temp': ev,
        'structural_integrity': si,
        'tank_pressure': tp,
        'critical_modules': cm,
    }

    # ── Individual status dictionary ─────────────────────────

    telemetry_status = {
        'internal_temp':        '[ OK ]'     if 21 <= iv <= 30  else '**FAILED**',
        'external_temp':        '[ OK ]'     if -5 <= ev <= 30  else '**FAILED**',
        'structural_integrity': '[ OK ]'     if si == 1         else '**FAILED**',
        'tank_pressure':        '[ OK ]'     if 95 <= tp <= 145 else '**FAILED**',
        'critical_modules':     '[ OK ]'     if cm == 1         else '**FAILED**',
    }

    return telemetry_data, telemetry_status
```

```python
def calculate_autonomy():
    total_capacity        = round(random.uniform(800, 1200), 1)   # total battery capacity
    current_charge_percent = round(random.uniform(60, 100), 1)   # current charge (%)
    takeoff_consumption_kw = round(random.uniform(15, 30), 1)    # estimated takeoff consumption
    energy_loss_perc       = round(random.uniform(1, 7), 1)      # energy losses (%)

    # ── Intermediate calculations ───────────────────────────────────────────────
    available_energy_kwh = round(total_capacity * (current_charge_percent / 100), 1)
    loss_kwh             = round(available_energy_kwh * (energy_loss_perc / 100), 2)
    usable_energy_kwh    = round(available_energy_kwh - loss_kwh, 4)
    autonomy_months      = round(random.uniform(3, 15))  # estimated autonomy in months
    autonomy_hours       = round(autonomy_months * 30 * 24)  # converting months to hours

    autonomy_data = {
        'total_capacity':        total_capacity,
        'available_energy_kwh':  available_energy_kwh,
        'current_charge_percent': current_charge_percent,
        'takeoff_consumption_kw': takeoff_consumption_kw,
        'energy_loss_perc':      energy_loss_perc,
        'usable_energy_kwh':     usable_energy_kwh,
        'autonomy_months':       autonomy_months,
        'autonomy_hours':        autonomy_hours,
    }

    autonomy_status = {
        'Battery':          '[ OK ]'  if current_charge_percent >= 80  else 'INSUFFICIENT',
        'Required energy':  '[ OK ]'  if usable_energy_kwh      >= 10  else 'INSUFFICIENT',
        'Energy loss':      '[ OK ]'  if energy_loss_perc       <= 5   else 'CRITICAL',
        'Autonomy':         '[ OK ]'  if autonomy_months        >= 10  else 'INSUFFICIENT',
    }

    return autonomy_data, autonomy_status
```

```python
# ── Calling the functions and retrieving data ─────────────────────────────────────────
autonomy_data, autonomy_status = calculate_autonomy()
telemetry_data, telemetry_status = generate_telemetry()

# ── Launch decision ──────────────────

decision = 'READY FOR TAKEOFF' if all(
    s == '[ OK ]' for s in {**telemetry_status, **autonomy_status}.values()
) else 'TAKEOFF ABORTED'
```

```python
# --- Printing results in an organized format ------------------------

print("Initializing Aurora-Siger Project")
print("...\n...")
print(input("Press Enter to start the analyses "))
print("Calculating battery autonomy...")
print(f"Processing {random.randint(25, 150)} autonomy calculations...")
input("Press Enter to continue...")
print("Generating telemetry calculations...")
print(f"{random.randint(25, 150)} telemetry records processed...")
input("Press Enter to continue...")

print("\n\n           === AUTONOMY DATA ===\n")
print(f"    Total battery capacity:             {autonomy_data['total_capacity']}kW")
print(f"    Current battery charge:             {autonomy_data['available_energy_kwh']}kW ({autonomy_data['current_charge_percent']}%)")
print(f"    Estimated takeoff consumption:      {autonomy_data['takeoff_consumption_kw']}kW")
print(f"    Energy loss:                        {autonomy_data['energy_loss_perc']}%")
print(f"    Battery after takeoff:              {autonomy_data['autonomy_hours']} hours ({autonomy_data['autonomy_months']} months)")

print("\n\n           === TELEMETRY DATA ===\n")
print(f"    Internal temperature:               {telemetry_data['internal_temp']}ºC")
print(f"    External temperature:               {telemetry_data['external_temp']}ºC")
print(f"    Structural integrity:               {telemetry_data['structural_integrity']}")
print(f"    Tank pressure:                      {telemetry_data['tank_pressure']}bar")
print(f"    Critical modules:                   {telemetry_data['critical_modules']}")

print("\n\n           === AUTONOMY STATUS ===\n")
print(f"    Battery:            {autonomy_status['Battery']}")
print(f"    Required energy:    {autonomy_status['Required energy']}")
print(f"    Energy loss:        {autonomy_status['Energy loss']}")
print(f"    Autonomy:           {autonomy_status['Autonomy']}")

print("\n\n           === TELEMETRY STATUS ===\n")
print(f"    Internal temperature:       {telemetry_status['internal_temp']}")
print(f"    External temperature:       {telemetry_status['external_temp']}")
print(f"    Structural integrity:       {telemetry_status['structural_integrity']}")
print(f"    Tank pressure:              {telemetry_status['tank_pressure']}")
print(f"    Critical modules:           {telemetry_status['critical_modules']}")

print(f"\nLaunch decision: {decision}")
```
